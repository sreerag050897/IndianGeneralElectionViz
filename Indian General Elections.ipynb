```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'E:\Election.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop('The largest party',axis=1,inplace=True)
```

```python
df['Election year']=df['Election year'].apply(lambda x: x.split('–')[0])
```

```python
df
```

```python
df['Total seats'] = df['Total seats'].ffill()
```

```python
my_map = {'Turnout': 'Turnout%', 'Percentage in the Lok Sabha': 'LS seat %'}
```

```python
df.rename(my_map,axis=1,inplace=True)
```

```python
df['Turnout%'] = df['Turnout%']*100
df['LS seat %'] = df['LS seat %']*100
```

```python
df
```

```python
df['Margin of majority'] = df['Margin of majority'].str.replace('–','-')
```

```python
df['Margin of majority'] = df['Margin of majority'].astype('Int64')
```

```python
df['Margin of majority'].dtype
```

```python
def Majority(a):
    if a>0:
        return "Majority"
    else :
        return "Hung"
```

```python
df['Gov_Status'] = df['Margin of majority'].apply(Majority)
```

```python
df['Ordinal Num of LS'] = range(1,len(df)+1)
```

```python
list(df.columns)
```

```python
df= df.reindex(columns=['Election year',
 'Lok Sabha', 'Ordinal Num of LS',
 'Total seats',
 'Turnout%',
 'The largest party_1',
 'Seats won by the largest party',
 'Margin of majority',
 'LS seat %',
 'Leader',
 'Gov_Status'
])
```

```python
df.drop('Lok Sabha',axis=1,inplace=True)
```

```python
df['Election year'] = df['Election year'].astype('Int64')
```

```python
df['Total seats'] = df['Total seats'].astype('Int32')
```

```python
df
```

```python
df.iloc[12,4] = 'Bharatiya Janata Party'
```

```python
df
```

```python
palette_color={'Indian National Congress':'Blue','Bharatiya Janata Party':'darkorange','Janata Party':'lime','Janata Dal':'green'}
```

```python
plt.figure(figsize=(14,8))
plt.axhline(y=272,color='red',linestyle='--',linewidth=1.5)
x=sns.barplot(data=df,x='Election year',y='Seats won by the largest party',hue='The largest party_1',dodge=False,palette=palette_color)
```

```python
plt.figure(figsize=(10, 8))
plt.axhline(0, color='black',linewidth=0.5)
k=plt.bar(df["Election year"], df["Margin of majority"], color=['green' if x > 0 else 'red' for x in df["Margin of majority"]])
plt.xlabel('Election Year')
plt.ylabel('Margin of Majority')
plt.title('Margin of Majority in Indian General Elections')
#plt.show()
plt.savefig('Majority.png', dpi=300, transparent=True)
```
